```python
df['Month'] = df['tpep_pickup_datetime'].dt.month_name()
df['Day'] = df['tpep_pickup_datetime'].dt.day_name()
df.head(10)
```

```python
month_rides = df['Month'].value_counts()
month_rides
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month_rides = month_rides.reindex(index=month_order)
month_rides
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x=month_rides.index,y=month_rides,hue=month_rides.index)
ax.set_xticks(month_order,labels=month_order,fontsize=9)
plt.ylabel('Frequency')
ax.set_title('Monthly Rides',fontsize=15)
plt.show()
```

```python
day_rides = pd.Series(df['Day']).value_counts()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday', 'Saturday', 'Sunday']
day_rides = day_rides.reindex(index=day_order)
day_rides
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x=day_order,y=day_rides,hue=day_order)
ax.set_xticks(day_order,labels=day_order)
plt.xlabel('Day')
plt.ylabel('Frequency')
ax.set_title('Daily Rides', fontsize=15)
plt.show()
```

```python
day_total = df.groupby(by='Day')['total_amount'].sum()
day_total = day_total.reindex(index=day_order)
day_total
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x=day_order,y=day_total,hue=day_order,palette='magma')
plt.xlabel('Day')
plt.ylabel('Revenue')
ax.set_xticks(day_order,labels=day_order)
ax.set_title('Daily Revenue',fontsize=15)
plt.show()
```

```python
month_total = df.groupby(by='Month')['total_amount'].sum()
month_total = month_total.reindex(index=month_order)
month_total
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x=month_order,y=month_total,hue=month_order,palette='magma')
ax.set_xticks(month_order,labels=month_order,fontsize=9)
plt.xlabel('Month')
plt.ylabel('Revenue')
ax.set_title('Monthly Revenue',fontsize=15)
plt.show()
```

```python
test = np.round(np.random.normal(10,5,(3000,2)),1)
mid = int(len(test)/2)
start = test[:mid]
end = test[mid:]
xpoint = np.array([start])
ypoint = np.array([end])
plt.scatter(xpoint,ypoint)
plt.show()
```

```python
dist = np.sqrt(((start-end)**2).sum(axis=-1))
```

```python
test_df = pd.DataFrame({'start':[tuple(x) for x in  start.tolist()],'end':[tuple(x) for x in end.tolist()],'distance':dist})
data2 = test_df[['end','distance']].groupby(['end']).mean()
data2 = data2.sort_values(by='distance')
test_df
```

```python
data2.reset_index(inplace=True)
data2
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x=data2.index,y=data2.distance,order=data2.index,color='white',edgecolor=None)
ax.set_xticks([])
plt.xlabel('Endpoints')
plt.ylabel('Mean Distance to all Other Points')
plt.title('Mean Distance between points taken randomly from a Normal Distribution',fontsize=15)
plt.show()
```
